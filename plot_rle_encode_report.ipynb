```python
from itertools import islice
import csv
import matplotlib.pyplot as plt
```

```python
report_file = open("rle_encode_report.csv")
pixel_counts = []
serial_durations = []
cuda_durations = []
omp_durations = []
thread_durations = []

for row in islice(csv.reader(report_file), 1, None):
    [pixel_count, serial_duration, cuda_duration, omp_duration, thread_duration] = row
    pixel_counts.append(int(pixel_count))
    serial_durations.append(int(serial_duration))
    cuda_durations.append(int(cuda_duration))
    omp_durations.append(int(omp_duration))
    thread_durations.append(int(thread_duration))

plt.plot(pixel_counts, serial_durations)
plt.plot(pixel_counts, cuda_durations)
plt.plot(pixel_counts, omp_durations)
plt.plot(pixel_counts, thread_durations)
plt.title('Time taken to RLE encode an image')
plt.ylabel('Duration (ms)')
plt.xlabel('Number of pixels')
plt.legend(['Serial', 'CUDA', 'OpenMP', 'std::thread'], loc='upper left')
plt.show()
plt.close()
```

```python
report_file.close()
```

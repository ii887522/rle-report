```python
from itertools import islice
import csv
import matplotlib.pyplot as plt
```

```python
report_file = open("rle_decode_colorful_cuda.csv")
image_sizes = []
serial_durations = []
cuda_128_durations = []
cuda_256_durations = []
cuda_512_durations = []

for row in islice(csv.reader(report_file), 1, None):
    [image_size, serial_duration, cuda_128_duration, cuda_256_duration, cuda_512_duration] = row
    image_sizes.append(int(image_size))
    serial_durations.append(float(serial_duration))
    cuda_128_durations.append(float(cuda_128_duration))
    cuda_256_durations.append(float(cuda_256_duration))
    cuda_512_durations.append(float(cuda_512_duration))

plt.plot(image_sizes, serial_durations)
plt.plot(image_sizes, cuda_128_durations)
plt.plot(image_sizes, cuda_256_durations)
plt.plot(image_sizes, cuda_512_durations)
plt.title('Time taken to RLE decode\n(Lower is better)')
plt.ylabel('Execution time (seconds)')
plt.xlabel('Image size (number of pixels on each side)')
plt.legend(['Serial', 'CUDA 128 threads per block', 'CUDA 256 threads per block', 'CUDA 512 threads per block'], loc='upper left')
plt.show()
plt.close()
```

```python
report_file.close()
```
